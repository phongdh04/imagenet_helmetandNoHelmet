```python
import os

# Mount Google Drive
from google.colab import drive
```

```python
# Đường dẫn tới thư mục ảnh trong Drive
folder_path = '/content/gdrive/MyDrive/dataset/Clear_Helmet'

# Các định dạng ảnh cần đếm
image_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.gif']
```

```python
# Đếm số ảnh
count = 0
for file in os.listdir(folder_path):
    if os.path.splitext(file)[1].lower() in image_extensions:
        count += 1

print(f"Số lượng ảnh trong thư mục: {count}")
```
